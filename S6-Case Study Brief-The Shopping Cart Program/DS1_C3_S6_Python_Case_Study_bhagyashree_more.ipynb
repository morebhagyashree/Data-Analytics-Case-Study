```python
total_cost=0.0
biscuit=cereals=chicken=oats=rice=0
t1=t2=t3=t4=t5=dc=0
q1=5
q2=10
q3=20
q4=20
q5=30
c1=2.50
c2=3.50
c3=13.30
c4=3.90
c5=4.50
name=""
address=""
dis=0
def display(biscuit,cereals,chicken,oats,rice):
    print("%60s" %("ESSENTIAL ITEMS AVAILABLE"))
    print("%70s" %("-------------------------------------------"))
    print("%20s %20s %20s %20s" %("S.No.","Item","Quantity","Cost/Item"))

    print("%20s %30s %7s %17s %1s" %("1.","Biscuits: a pack of 20",q1-biscuit,"$",c1))
    print("%20s %30s %7s %17s %1s" %("2.","Cereals:2 pounds/pack",q2-cereals,"$",c2))
    print("%20s %30s %7s %17s %1s" %("3.","Chicken: 1 pound/pack",q3-chicken,"$",c3))
    print("%20s %30s %7s %17s %1s" %("4.","Oats:4 pounds/pack",q4-oats,"$",c4))
    print("%20s %30s %7s %17s %1s" %("5.","Rice:2 pounds/pack",q5-rice,"$",c5))

def display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address):
    print("%60s" %("BILL"))
    print("%70s" %("-------------------------------------------"))
    print("%20s %20s %20s %20s" %("S.No.","Item","Quantity","Total Cost"))

    print("%20s %30s %7s %17s %1s" %("1.","Biscuits",biscuit,"$",c1*biscuit))
    print("%20s %30s %7s %17s %1s" %("2.","Cereals",cereals,"$",c2*cereals))
    print("%20s %30s %7s %17s %1s" %("3.","Chicken",chicken,"$",c3*chicken))
    print("%20s %30s %7s %17s %1s" %("4.","Oats",oats,"$",c4*oats))
    print("%20s %30s %7s %17s %1s" %("5.","Rice",rice,"$",c5*rice))
    print("Total Item Cost: ",total_cost)
    print("Total Bill amount: Total Item Cost + Delivery Charge =",total_cost+dc)
    print("Name: ",name)
    print("Address: ",address)
    
def userinput(total_cost,c1,c2,c3,c4,c5,dc):
  
    biscuit=int(input("How many packets of biscuits are required?  "))
    cereals=int(input("How many packets of cereals are required?  "))
    chicken=int(input("How many packets of chicken are required?  "))
    oats=int(input("How many packets of oats are required?  "))
    rice=int(input("How many packets of rice are required?  "))
    print("Fill in your delivery details")
    name=input("Enter your name ")
    address=input("Enter your address ")
    dis=int(input("Distance from store 0-5/5-10/10-15/Above 15 "))
   
    
    if(dis<5):
        dc=10
    elif(dis>=5 and dis<10):
        dc=12
    elif(dis>=10 and dis<15):
        dc=14
    else:
        dc=20
    
    
    if(biscuit>=q1):
        c1=0
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        print("biscuit=====",biscuit)
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        print("Biscuit Item Out of Stock")
        
        
    elif(cereals>=q2):
        c2=0
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        print("Cereals Item Out of Stock")
        
        

    elif(chicken>=q3):
        c3=0
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        print("Chicken Item Out of Stock")
        
       


    elif(oats>=q4):
        c4=0
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        print("Oats Item Out of Stock")
       

    elif(rice>=q5):
        c5=0
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        print("Rice Item Out of Stock")
        
    
    else:
        total_cost=(biscuit*c1)+(cereals*c2)+(chicken*c3)+(oats*c4)+(rice*c5)
        display_cost(total_cost,biscuit,cereals,chicken,oats,rice,c1,c2,c3,c4,c5,dc,name,address)
        
    while True:
        choice = input("Do you want to continue shopping? (y/n): ")
        if choice.lower() == "y":
            
            display(biscuit,cereals,chicken,oats,rice)
            userinput(total_cost,c1,c2,c3,c4,c5,dc)
        elif choice.lower() == "n":
            print("Thank you for shopping with us.....")
            break
        else:
            print("Invalid input. Please enter 'y' or 'n'.")


display(biscuit,cereals,chicken,oats,rice)
userinput(total_cost,c1,c2,c3,c4,c5,dc)



```

Initialize the variables for total_cost, biscuit, cereals, chicken, oats, rice, t1, t2, t3, t4, t5, dc, q1, q2, q3, q4, q5, c1, c2, c3, c4, c5, name, address, and dis.
Define the display(biscuit, cereals, chicken, oats, rice) function:
a. Print the header and separator lines for the available items.
b. Print the item details for each item, including the serial number, item name, available quantity, and cost per item.
Define the display_cost(total_cost, biscuit, cereals, chicken, oats, rice, c1, c2, c3, c4, c5, dc, name, address) function:
a. Print the header and separator lines for the bill.
b. Print the item details for each item in the bill, including the serial number, item name, purchased quantity, and total cost.
c. Print the total item cost, total bill amount (including delivery charge), name, and address.
Define the userinput(total_cost, c1, c2, c3, c4, c5, dc) function:
a. Prompt the user to input the quantities of biscuits, cereals, chicken, oats, and rice required.
b. Prompt the user to input their name, address, and distance from the store.
c. Determine the delivery charge based on the distance.
d. Check if any item is out of stock based on the entered quantities.
e. Calculate the total cost based on the entered quantities and prices.
f. Call the display_cost() function to display the bill with item quantities and costs.
g. Use a loop to ask the user if they want to continue shopping.
i. If the user wants to continue, call the display() function to show the available items.
ii. Recursively call the userinput() function to get new input and calculate the new total cost.
iii. If the user does not want to continue, end the program with a "Thank you" message.
